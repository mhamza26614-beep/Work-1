```python
Value1 = float(input("Enter any first number:"))
Value2 = float(input("Enter any second number:"))
Operation = input("Enter the Operation (+,-,*,/:")
if Operation == "+":
    print("Result", Value1 + Value2)
elif Operation == "-":
    print("Result", Value1 - Value2)
elif Operation == "*":
    print("Result", Value1 * Value2)
elif Operation == "/":
    if Value2 != 0:
        print("Result", Value1 / Value2)
    else:
        print("Can't divide by Zero")
elif Operation == "**":
    print("Result", Value1 ** Value2)
else:
    print("Invalid Operation!")
```

```python
my_list = ["apple", "banana", 10, "orange"]

found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("There is at least one numeric value in the list.")
else:
    print("There is no numeric value in the list.")
```

```python
my_dict = {
    "a": 10,
    "b": 20,
    "c": "Python",
    "d": 5.5
}
total = 0
for value in my_dict.values():
    if isinstance(value, (int, float)):
        total += value
print("Sum of numeric values:", total)
```

```python
my_list = [1, 2, 3, 2, 4, 5, 1, 6]

seen = set()
duplicates = set()

for item in my_list:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)
print("Duplicate values:", list(duplicates))
```
