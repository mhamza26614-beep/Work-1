```python
Value1 = float(input("Enter any first number:"))
Value2 = float(input("Enter any second number:"))
Operation = input("Enter the Operation (+,-,*,/:")
if Operation == "+":
    print("Result", Value1 + Value2)
elif Operation == "-":
    print("Result", Value1 - Value2)
elif Operation == "*":
    print("Result", Value1 * Value2)
elif Operation == "/":
    if Value2 != 0:
        print("Result", Value1 / Value2)
    else:
        print("Can't divide by Zero")
else:
    print("Invalid Operation!")
```

```python
Eng = int(input ("Enter english marks:"))
Maths = int(input ("Enter maths marks:"))
Urdu = int(input ("Enter urdu marks:"))
Comp = int(input ("Enter computer marks:"))
Isl = int(input ("Enter islamiat marks:"))
totalMarks = Eng+Isl+Maths+Urdu+Comp
print("Your Total Marks is:", totalMarks)
perc = totalMarks/500*100
print ("Your Percentage is:",perc)
if perc>= 90 and perc<= 100:
   print("A* Grade")
elif perc>= 80 and perc<= 90:
   print("A Grade")
elif perc>= 70 and perc<= 80:
   print("B Grade")
elif perc>= 60 and perc<= 70:
   print("C Grade")
elif perc>= 50 and perc<= 60:
   print("D Grade")
elif perc>= 34 and perc<= 50:
   print("E grade")
elif perc>= 25 and perc<= 33:
   print("Fail")
```
